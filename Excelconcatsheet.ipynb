```python
import pandas as pd
from google.colab import files

# Mengunggah file Excel ke sesi Colab
uploaded = files.upload()

# Nama sheet yang akan digabungkan
sheet_name = 'Kas'

# List untuk menampung data dari setiap file
data_frames = []

# Melakukan iterasi pada setiap file yang diunggah
for file_name, content in uploaded.items():
    # Membaca file Excel dan mengambil sheet yang diinginkan
    data = pd.read_excel(content, sheet_name=sheet_name)
    
    # Menambahkan DataFrame ke list
    data_frames.append(data)

# Menggabungkan DataFrame dari list menjadi satu DataFrame
combined_data = pd.concat(data_frames, ignore_index=True)

# Menyimpan data gabungan menjadi file Excel
combined_data.to_excel('gabungan.xlsx', index=False)

print("Data berhasil digabungkan dan disimpan dalam file 'gabungan.xlsx'")
files.download('gabungan.xlsx')
```
